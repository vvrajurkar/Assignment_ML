### 2 Create the below pattern using nested for loop in Python

```python
max = int(input("Enter maximum number of stars you need in pattern: "))
```

```python
for i in range(0,max):
    for j in range(0,max):
        if(j<=i):
            print("*",end=" ") 
        else:
            print()
            break
print()
for i in range(max,0,-1):
    for j in range(0,max):
        if(j<i-1):
            print("*",end=" ")
        else:
            print()
            break
```
